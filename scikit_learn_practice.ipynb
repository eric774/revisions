Problem Statement

You have been provided with a dataset that contains the costs of advertising on different
media channels and the corresponding sales of XYZ firm. Evaluate the dataset to:

1. Find the features or media channels used by the firm
2. Find the sales figures for each channel

3. Create a model to predict the sales outcome

4. split it into training and testing datasets for the model
5. Calculate the mean squared error (MSE)

```python
import pandas as pd
path = '/content/Advertising_Budget_and_Sales.csv'
df_adv_data = pd.read_csv(path, index_col = 0)
df_adv_data.head()
```

```python
print(df_adv_data.size)
print(df_adv_data.shape)
df_adv_data.columns = ['TV', 'Radio', 'Newspaper', 'Sales']
print(df_adv_data.columns)
```

```python
X_feature = df_adv_data[['Newspaper', 'Radio', 'TV']]
X_feature.head()
```

```python
Y_target = df_adv_data[['Sales']]
Y_target.head()
```

```python
#split test and training data
# by default 75% training data and 25% testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_feature,Y_target, random_state=1)
```

```python
#view shape of train and test data sets for both feature and response
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Linear regression model
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train,y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
y_pred = linreg.predict(x_test)
print(y_pred)
```

```python
from sklearn import metrics
import numpy as np
#mse
print(np.sqrt(metrics.mean_squared_error(y_pred, y_test)))
```
